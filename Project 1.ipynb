```python
heart_data = pd.read_json("heart.json") #read json file
```

```python
modified_heart = heart_data.drop(['target'], axis = 1) #remove last column of table
modified_heart.head()
```

```python
columns = modified_heart.shape[0] 
records = modified_heart.shape[1] #rows
print("Data set contains", columns, "columns and", records, "records.") 
```

```python
# Finding the health status of a serum cholesterol (mg/dL) value as the input, should output the health status in addition to all data values in the data set that have the specified cholesterol value. 
def health_status(chol):
    if chol in modified_heart['chol'].unique():
        if chol < 125:
            data_sets = modified_heart.loc[modified_heart['chol'] == chol]
            print("Low cholesterol level") 
        if chol > 200:
            data_sets = modified_heart.loc[modified_heart['chol'] == chol]
            print("High cholesterol level")
        if 125 < chol < 200:
            data_sets = modified_heart.loc[modified_heart['chol'] == chol]
            print("Healthy cholesterol level")
        if type(chol) == None or type(chol) == string:
            print("Health status not available")
        return data_sets
    else:
        print("Health status not available")
```

```python
health_status(61)
```
